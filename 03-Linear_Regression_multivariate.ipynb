```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

Setup a two-dimensional datagrid

```python
N = 10
x1 = np.linspace(0,1,N)
x2 = np.linspace(0,1,N)
```

Iterate over x1 and x2 and fill out a mtrix of independent variables X and the corresponding output Y. We make also a scatterplot, where the circle size depends on y

```python
X = np.zeros((100,2))
Y = np.zeros((100,1))

for i in range(10):
    for j in range(10):
        row = 10*i+j
        X[row,0] = x1[i]
        X[row,1] = x2[j]
        Y[row,0] = 1.1*x1[i] - 0.7*x2[j]

# plt.plot(X[:,0],X[:,1],'rx')
plt.scatter(X[:,0],X[:,1],s=100*Y)
```

Now, let's make a linear regression and output the parameters. Note that we have a model $\hat y = c_1 x_1 + c_2 x_2 + d$ and search for the three parameters $c_1, c_2, d$.

```python
lin_reg = LinearRegression()
lin_reg.fit(X, Y)

print(lin_reg.intercept_, lin_reg.coef_)
```
